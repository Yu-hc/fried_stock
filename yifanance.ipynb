```python
import pandas as pd
from FinMind.data import DataLoader

# Initialize the DataLoader (optionally provide a user_id and password for API access if needed for premium features)
dl = DataLoader()

# Fetch daily Taiwan stock data for a specific stock (e.g., '2330' for TSMC)
stock_data = dl.taiwan_stock_daily(
    stock_id='2330',  # Stock symbol
    start_date='2023-01-01',  # Start date in YYYY-MM-DD format
    end_date='2023-12-31'     # End date in YYYY-MM-DD format
)

# Print the first few rows of the data (it's a pandas DataFrame)
print(stock_data.head())

# Add additional features, like institutional investor data
stock_data = dl.feature.add_kline_institutional_investors(stock_data)

# Save to CSV if needed
stock_data.to_csv('stock_data.csv', index=False)

# Optional: Plot a K-line (candlestick) chart if you have matplotlib installed
from FinMind import plotting
plotting.kline(stock_data)
```
